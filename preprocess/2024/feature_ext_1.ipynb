```python
def find_top_incomes(df, top_n=3):
    # 筛选出正数的收入记录
    df_income = df[df['ACTG_DIRET_CD'] == '1250d7cb654a81c7b9366dabf57fe62b'].copy()

    # 对每个企业进行分组，找出前几笔最大的收入
    df_income['rank'] = df_income.groupby('CUST_NO')['RMB_TR_AMT'].rank(method='first', ascending=False)
    df_top_incomes = df_income[df_income['rank'] <= top_n]

    return df_top_incomes
```

```python
import pandas as pd
from itertools import combinations
from functools import reduce


def expand_dataframe(df, values, index, columns, aggfunc):
    """
    展开DataFrame的维度，接收values，index，columns，aggfunc四个参数，
    对columns进行组合，依次对这些组合进行pivot_table，最后把结果合并起来。

    参数：
    - df: 原始DataFrame
    - values: 要聚合的值的列名列表
    - index: 用作索引的列名列表
    - columns: 要进行组合的列名列表
    - aggfunc: 聚合函数列表

    返回：
    - 合并后的DataFrame，包含所有列组合的pivot_table结果
    """
    pivot_tables = []
    # 生成columns的所有组合，组合长度从1到len(columns)
    for r in range(1, len(columns) + 1):
        combos = list(combinations(columns, r))
        for cols in combos:
            # 创建pivot_table
            pivot = df.pivot_table(
                values=values,
                index=index,
                columns=list(cols),
                aggfunc=aggfunc,
                fill_value=0
            )
            # 处理列名，避免冲突
            pivot.columns = ['_'.join([str(i) for i in col]).strip() for col in pivot.columns.values]
            # 添加列组合的标识
            pivot = pivot.rename(columns=lambda x: f"{x}_{'_'.join(cols)}")
            # 重置索引，方便后续合并
            pivot = pivot.reset_index()
            pivot_tables.append(pivot)

    # 合并所有pivot_table的结果
    result = reduce(lambda left, right: pd.merge(left, right, on=index, how='outer'), pivot_tables)

    return result


# 展开！
result = expand_dataframe(
    df=df_merged,
    values=['RMB_TR_AMT', 'ARG_ACCT_BAL'],
    index=['CUST_NO'],
    columns=['ACTG_DIRET_CD', 'TRS_CSH_IND', 'CSH_EX_IND', 'INT_BNK_TR_IND', 'CPT_INTL_FE_CUST_IND', 'SAME_NAM_IND'],
    aggfunc=['sum']
)


```
