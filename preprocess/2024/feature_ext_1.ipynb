```python
def find_top_incomes(df, top_n=3):
    # 筛选出正数的收入记录
    df_income = df[df['RMB_TR_AMT'] > 0].copy()
    
    # 对每个企业进行分组，找出前几笔最大的收入
    df_income['rank'] = df_income.groupby('CUST_NO')['RMB_TR_AMT'].rank(method='first', ascending=False)
    df_top_incomes = df_income[df_income['rank'] <= top_n]
    
    return df_top_incomes

# 假设已经加载并合并了金融性交易明细数据
df_fncl_tr_dtal = loader.to_concat_df('XW_ENTINFO_FNCL_TR_DTAL')

# 找到每个企业的前3笔最大收入
df_top_incomes = find_top_incomes(df_fncl_tr_dtal, top_n=1)
df_top_incomes.head()
```
